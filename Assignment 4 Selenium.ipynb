Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Series
B) Place
C) Date
D) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
driver=webdriver.Chrome()
driver.get("https://www.bcci.tv/")
```

```python
f=driver.find_element(By.XPATH,'/html/body/header/div[3]/div[2]/ul/div[1]/a[2]')
f.click()
time.sleep(20)
driver.find_element(By.XPATH,'/html/body/section/div/div/div/div/div/div[2]/div[2]/div[2]/div/button').click()
```

```python
series=[]
place=[]
date=[]
time=[]
```

```python
teams=driver.find_elements(By.XPATH,'//h5[@class="match-tournament-name ng-binding"]')
for i in teams:
    if i.text is None:
        series.append('Not')
    else:
        series.append(i.text)
match_place=driver.find_elements(By.XPATH,'//div[@class="match-place ng-scope"]')
for i in match_place:
    if i.text is None:
        place.append('Not')
    else:
        place.append(i.text)
match_date=driver.find_elements(By.XPATH,'//div[@class="match-dates ng-binding"]') 
for i in match_date:
    if i.text is None:
        date.append('Not')
    else:
        date.append(i.text)
match_time=driver.find_elements(By.XPATH,'//div[@class="match-time no-margin ng-binding"]')
for i in match_time:
    if i.text is None:
        time.append('Not')
    else:
        time.append(i.text)
```

```python
len(series),len(place),len(date),len(time)
```

```python
df=pd.DataFrame({"Series":series,"Place":place,"Date":date,"Time":time})
df
```

8. Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/ You
have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute G) Year
Note: - from the home page you have to go to the Show All Dataset page through code.

```python
driver=webdriver.Chrome()
driver.get("https://archive.ics.uci.edu/")
```

```python
driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/div/div[2]/button').click()
dataset=driver.find_element(By.XPATH,'//a[@class="btn-primary btn"]')
dataset.click()
```

```python
g=driver.find_element(By.XPATH,'//label[@class="btn-primary btn-sm btn flex gap-2 rounded-full"]').click()
```

```python
Dataset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
No_of_instances=[]
No_of_attribute=[]
Year=[]
start=0
end=3
for r in range(start,end):
    d=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
    for i in d:
        Dataset_name.append(i.text)
    y=driver.find_elements(By.XPATH,'//span[@class="lister-item-year text-muted unbold"]')   

    g=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[1]')   
    for i in g:
        Task.append(i.text)  
    g=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[2]')   
    for i in g:
        Data_type.append(i.text)  
    Data_type     
    g=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[3]')   
    for i in g:
        No_of_instances.append(i.text)  
    No_of_instances 
    g=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[4]')   
    for i in g:
        No_of_attribute.append(i.text)  
    g=driver.find_elements(By.XPATH,'//tbody//tr//td[2]')
    for i in g:
        Attribute_type.append(i.text)
    g=driver.find_elements(By.XPATH,'//tbody//tr//td[3]')
    for i in g:
        Year.append(i.text)
    Year  
next_btn=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]').click()
```

```python
from datetime import datetime
```

```python
years=[]
for date_str in Year:
    try:
        date_object = datetime.strptime(date_str, '%m/%d/%Y')
        year = date_object.year
        years.append(year)
    except ValueError:
        years.append('N/A')
        
years        
```

```python
Dataset_details = pd.DataFrame({})
Dataset_details['Name']= Dataset_name
Dataset_details['Data Type'] = Data_type
Dataset_details['Task'] = Task 
Dataset_details['Attribute type'] = Attribute_type
Dataset_details['Attribute Number'] = No_of_attribute
Dataset_details['Instances Number'] = No_of_instances
Dataset_details['Years'] = years 
Dataset_details
```
